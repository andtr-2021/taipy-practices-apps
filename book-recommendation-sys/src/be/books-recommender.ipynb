```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import plotly.express as px
warnings.filterwarnings("ignore")
```

```python
book_df=pd.read_csv("../../data/data4display/goodreads-books.csv")
book_df.head(5)
```

```python
book_df = book_df.dropna(subset=['genres'])
book_df["genres"]=book_df["genres"].apply(lambda x: ';'.join( x.split(';')[:4] ).split(",")[0] )
```

```python
book_df.isnull().sum()
```

```python
book_df.Title.duplicated().sum()
book_df.drop_duplicates(subset='Title', keep='first', inplace=True)
```

```python
data=book_df.copy()
```

```python
data.loc[(data["average_rating"]>=0) & (data["average_rating"]<=1),"rating_between"]="between_0_to_1"
data.loc[(data["average_rating"]>1) & (data["average_rating"]<=2),"rating_between"]="between_1_to_2"
data.loc[(data["average_rating"]>2) & (data["average_rating"]<=3),"rating_between"]="between_2_to_3"
data.loc[(data["average_rating"]>3) & (data["average_rating"]<=4),"rating_between"]="between_3_to_4"
data.loc[(data["average_rating"]>4) & (data["average_rating"]<=5),"rating_between"]="between_4_to_5"
data.head()
```

```python
between_df=data["rating_between"].str.get_dummies(",")
between_df
```

```python
language_df=data["language_code"].str.get_dummies(",")
language_df.head()
```

```python
genres_df=data["genres"].str.get_dummies(";")
genres_df.head()
```

```python
engine_features=pd.concat([between_df,language_df,genres_df,data["average_rating"],data["ratings_count"]],axis=1)
engine_features
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
engine_features = min_max_scaler.fit_transform(engine_features)
```

```python
from sklearn import neighbors
engine_model=neighbors.NearestNeighbors(n_neighbors=6,algorithm="ball_tree") # KD-Tree,Cover Tree,Brute-Force,Gaussian Mixture Models
engine_model.fit(engine_features)
dist, idlist = engine_model.kneighbors(engine_features)
```

```python
def book_recommendation_engine(book_name):
    book_list_name = []
    book_list_genre= []
    book_id = data[data['Title'] == book_name].index
    if book_id.size == 0:
        print("Book not found!")
        return book_list_name,book_list_genre
    for newid in idlist[book_id]:
        book_list_name.append(data.iloc[newid].Title)
        for index, row in genres_df.iterrows():
            genre = row[row == 1].index.values  
            book_list_genre.append(list(genre))
    return book_list_name,book_list_genre
```

```python
from colorama import Fore, Style

name = "Bill Bryson's African Diary"

book_list_name,book_list_genre = book_recommendation_engine(name)

if len(book_list_name) > 0:
    for i in range(len(book_list_name)):
        a=str(list(book_list_genre[i])).replace("[","").replace("]","").replace("'","")
        print("recommediton name: \n",book_list_name[i],"  genres: ", Fore.BLUE + a + Style.RESET_ALL)

```
